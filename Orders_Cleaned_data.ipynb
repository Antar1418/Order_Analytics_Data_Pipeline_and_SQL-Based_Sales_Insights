```python
import pandas as pd
```

```python
#Read data from the file and handle null values
df = pd.read_csv('Order_Py+SQL.csv', na_values=['Not Available','unknown'])
df.head(10)

# na_values=['Not Available','unknown'] This convert from 'Not Available','unknown' to nan(null)
```

```python
#Handling null values
df['Ship Mode'].unique()

#array(['Second Class', 'Standard Class', 'Not Available', 'unknown',
#       'First Class', nan, 'Same Day'], dtype=object)
```

```python
#rename columns names ..make them lower case and replace space with underscore
df.columns
```

```python
#df.rename(columns={'Order Id':'order id', 'Order Date':'order date'}) for single columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = df.columns.str.replace(' ','_')
```

# derive new columns discount , sale price and profit

```python
df.head(1)
```

```python
df['discount'] = df['list_price'] * df['discount_percent']/100  
```

```python
df.head(2)
```

```python
# Sale Price 
df['sale_price'] = df['list_price'] - df['discount']
df.head(5)
```

```python
#profit
df['profit'] = df['sale_price']-df['cost_price']
df.head(5)
```

# convert order date from object data type to datetime
